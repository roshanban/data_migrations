```python
import pandas as pd
```

```python
folder_path = r'F:\smapledata\heart_disease_raw.csv'
heart_df = pd.read_csv(folder_path)
heart_df.head()
```

```python
from sqlalchemy import create_engine
from sqlalchemy.types import Date

pg_engine = create_engine(
    'postgresql+psycopg2://postgres:test@localhost:5432/hell'
)
heart_df.to_sql(
    name="Heart_Disease_Track",
    con=pg_engine,
    if_exists="append",
    index=False,
    dtype={
        "file_date": Date
    }
)

print("Data successfully loaded into PostgreSQL")
```
